```python
import pandas as pd
import numpy as np
```

```python
data_dir = 'https://raw.githubusercontent.com/irl-labs/polis/main/notebooks/governance/finance/'
ISSUANCE_YEAR = '2023'
issuance = pd.read_csv(data_dir+'bond_issuance_2024.tsv',sep='\t')
trades   = pd.read_csv(data_dir+'bond_trades_20231101_20231231.tsv',sep='\t')

trades.datetime=pd.to_datetime(trades.datetime,format="%m/%d/%Y %I:%M:%S %p")
trades.maturity=pd.to_datetime(trades.maturity).dt.date
trades['maturity_year']=pd.to_datetime(trades.maturity).dt.year

mask   = (trades.trade_type=='Customer bought')&(trades.issuance_year==ISSUANCE_YEAR)
trades = trades[mask].reset_index(drop=True)

trades.price    = trades.price.astype(float)
trades['amount']= trades.quantity * trades.price / trades.unit

## exception - churn
mask = (trades.maturity_year==2033)&(trades.amount==29859.25)
trades = trades[~mask]
```

```python
summary = trades.groupby('maturity_year')\
    . agg({
        'unit':len,
        'amount':sum,
        'quantity':sum
    })\
    . reset_index()\
    . rename(columns={'maturity_year':'year','quantity':'trade_quantity','unit':'trades'})

max_duration_by_year = trades.groupby(pd.to_datetime(trades['maturity']).dt.year)['datetime'].apply(lambda x: x.max() - x.min())
summary['hours_to_complete'] = pd.Series(round(max_duration_by_year/ np.timedelta64(1, 'h'),1)).to_list()
summary=summary.merge(issuance,how='outer',on='year')

total_quantity_per_year = trades.groupby('maturity_year')['quantity'].sum()

trades['weight'] = trades.apply(lambda row: row['quantity'] / total_quantity_per_year[row['maturity_year']], axis=1)
trades['weighted_yield'] = trades['weight'] * trades['yield']
weighted_avg_yield_per_year = trades.groupby('maturity_year')['weighted_yield'].sum()

summary['trade_yield'] = pd.Series(round(weighted_avg_yield_per_year,2)).to_list()

summary['over_under']=summary['amount']-summary['principal']
assert((summary.trade_quantity!=summary.principal).any()==False)

summary['annual_interest']=summary['principal']*summary['coupon']/100
summary['total_interest']=(summary['principal']*summary['coupon']/100)*(summary['year']-int(ISSUANCE_YEAR))
```

```python
print(
    f"principal: ${summary.principal.sum():,.0f}\n"
    f"proceeds : ${summary.amount.sum():,.0f}\n"
    f"interest : ${summary.total_interest.sum():,.0f}\n"
    f"excess/under: ${summary.amount.sum()-summary.principal.sum():,.0f}\n\n"
)

print(
    f'interest rate: {100*(pow(1+((summary.principal.sum()+summary.total_interest.sum())/summary.principal.sum()),1/30)-1):.3f}%\n'
    f'avg   yield:   {summary["yield"].mean():.3f}%\n'
    f'trade yield:   {summary["trade_yield"].mean():.3f}%\n'
    f'weight coupon: {(summary.principal*summary["coupon"]).sum()/summary.principal.sum():.3f}%\n'
    f'avg coupon:    {(summary["coupon"]).mean():.3f}%\n\n'
)

print(
    f'$922,000 "cash from winning bidder"\nDealer expense: '
    f'{round(10000*(summary.amount.sum()-summary.principal.sum()-922000)/summary.principal.sum(),0):.0f} bps\n\n'
)

summary
```
